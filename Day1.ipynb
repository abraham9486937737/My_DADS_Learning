```python
import random

print("üéÆ Welcome to the Math Quiz Game!")

score = 0  # variable to track the score

for i in range(5):  # Loop through 5 questions
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    operation = random.choice(['+', '-', '*', '/', '%', '**'])

    # Avoid division by zero
    if operation == '/' and num2 == 0:
        continue

    # Forming question string
    question = f"{num1} {operation} {num2}"
    print(f"Question {i + 1}: What is {question}?")

    # Getting user input
    user_input = input("Your answer: ")

    # Check for empty input
    if user_input == "":
        print("You skipped the question.")
        pass
        continue

    # Convert input to float for decimal answers
    try:
        user_answer = float(user_input)
    except ValueError:
        print("‚ùå Invalid input. Skipping this question.")
        continue

    # Calculate correct answer
    if operation == '+':
        correct = num1 + num2
    elif operation == '-':
        correct = num1 - num2
    elif operation == '*':
        correct = num1 * num2
    elif operation == '/':
        correct = num1 / num2
    elif operation == '%':
        correct = num1 % num2
    elif operation == '**':
        correct = num1 ** num2

    # Comparing answers
    if user_answer == correct:
        print("‚úÖ Correct!")
        score += 1
    else:
        print(f"‚ùå Incorrect. The correct answer was {correct}")

    # Option to stop early
    if score == 3:
        print("üëè Great job! You scored 3/3. Ending early.")
        break

# Final Score
print(f"Your final score is {score}/5")

# Final message with if-else
if score == 5:
    print("üèÜ You're a Math Genius!")
elif score >= 3:
    print("üëç Good job! Keep practicing.")
else:
    print("üìò Review the basics and try again.")
```
